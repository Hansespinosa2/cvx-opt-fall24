```python
import matplotlib.pyplot as plt
plt.scatter([1], [1],c='green')
plt.scatter([1], [5],c='blue')
plt.scatter([1], [3],c='red')
plt.plot([0,2],[3,3],c='black')
plt.fill_between([0,2],[3,3],[0,0],color='black',alpha=0.3)
plt.legend('abc')
plt.axis('off')
plt.savefig('images/voronoi_halfspace.png')
plt.show()
```

```python
from mpl_toolkits.mplot3d import Axes3D

import matplotlib.pyplot as plt
import numpy as np

# Create a new figure and axis
fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')

# Define the coordinates
a_1 = np.array([1, 2, 3])
a_2 = np.array([4, 5, 2])
corner1 = a_1 + a_2
corner2 = a_1 + -1*a_2
corner3 = -1*a_1 + a_2
corner4 = -1*a_1 +  -1*a_2
corner5 = a_1
corner6 = a_2


# Plot the coordinates
ax.scatter(corner1[0], corner1[1], corner1[2], c='red')
ax.scatter(corner2[0], corner2[1], corner2[2], c='blue')
ax.scatter(corner3[0], corner3[1], corner3[2], c='green')
ax.scatter(corner4[0], corner4[1], corner4[2], c='yellow')
ax.scatter(corner5[0], corner5[1], corner5[2], c='black')
ax.scatter(corner6[0], corner6[1], corner6[2], c='black')


# Plot the lines
ax.plot([corner1[0], corner2[0]], [corner1[1], corner2[1]], [corner1[2], corner2[2]], c='black')
ax.plot([corner1[0], corner3[0]], [corner1[1], corner3[1]], [corner1[2], corner3[2]], c='black')
ax.plot([corner1[0], corner4[0]], [corner1[1], corner4[1]], [corner1[2], corner4[2]], c='black')
ax.plot([corner2[0], corner3[0]], [corner2[1], corner3[1]], [corner2[2], corner3[2]], c='black')
ax.plot([corner2[0], corner4[0]], [corner2[1], corner4[1]], [corner2[2], corner4[2]], c='black')
ax.plot([corner3[0], corner4[0]], [corner3[1], corner4[1]], [corner3[2], corner4[2]], c='black')


plt.legend(['a_1 + a_2', 'a_1 - a_2', '-a_1 + a_2', '-a_1 - a_2', 'a_1', 'a_2'])

# Set labels for the axes
ax.set_xlabel('X')
ax.set_ylabel('Y')
ax.set_zlabel('Z')

# Show the plot
plt.savefig('images/polyhedra_a.png')
plt.show()
```
