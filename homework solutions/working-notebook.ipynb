```python
from mpl_toolkits.mplot3d import Axes3D

import matplotlib.pyplot as plt
import numpy as np

# Create a new figure and axis
fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')

# Define the coordinates
a_1 = np.array([1, 2, 3])
a_2 = np.array([4, 5, 2])
corner1 = a_1 + a_2
corner2 = a_1 + -1*a_2
corner3 = -1*a_1 + a_2
corner4 = -1*a_1 +  -1*a_2
corner5 = a_1
corner6 = a_2


# Plot the coordinates
ax.scatter(corner1[0], corner1[1], corner1[2], c='red')
ax.scatter(corner2[0], corner2[1], corner2[2], c='blue')
ax.scatter(corner3[0], corner3[1], corner3[2], c='green')
ax.scatter(corner4[0], corner4[1], corner4[2], c='yellow')
ax.scatter(corner5[0], corner5[1], corner5[2], c='black')
ax.scatter(corner6[0], corner6[1], corner6[2], c='black')


# Plot the lines
ax.plot([corner1[0], corner2[0]], [corner1[1], corner2[1]], [corner1[2], corner2[2]], c='black')
ax.plot([corner1[0], corner3[0]], [corner1[1], corner3[1]], [corner1[2], corner3[2]], c='black')
ax.plot([corner1[0], corner4[0]], [corner1[1], corner4[1]], [corner1[2], corner4[2]], c='black')
ax.plot([corner2[0], corner3[0]], [corner2[1], corner3[1]], [corner2[2], corner3[2]], c='black')
ax.plot([corner2[0], corner4[0]], [corner2[1], corner4[1]], [corner2[2], corner4[2]], c='black')
ax.plot([corner3[0], corner4[0]], [corner3[1], corner4[1]], [corner3[2], corner4[2]], c='black')


plt.legend(['a_1 + a_2', 'a_1 - a_2', '-a_1 + a_2', '-a_1 - a_2', 'a_1', 'a_2'])

# Set labels for the axes
ax.set_xlabel('X')
ax.set_ylabel('Y')
ax.set_zlabel('Z')

# Show the plot
plt.savefig('hw1/images/polyhedra_a.png')
plt.show()
```

```python
import matplotlib.pyplot as plt



# Define the vectors
vectors = [[0, 0], [1, 0], [2/2.2, 1/2.2], [3/3.1, 1/3.1],[1/1.7, 1/1.7], [24/25,7/25]] 

# Extract x and y coordinates from the vectors
x = [vector[0] for vector in vectors]
y = [vector[1] for vector in vectors]

zeros_list = [0 for i in range(len(vectors))]

# Plot the vectors
plt.quiver(zeros_list, zeros_list, x, y, angles='xy', scale_units='xy', scale=1, color=['black', 'red', 'blue', 'green', 'orange'])

# Set labels for the axes
plt.xlabel('X')
plt.ylabel('Y')

# Set the aspect of the plot to be equal
plt.axis('equal')

plt.grid()

# Set the limits of the plot
plt.xlim(-1, 1)
plt.ylim(-1, 1)
plt.savefig('hw2/images/exercise233cone.png')
# Show the plot
plt.show()
```

```python
import numpy as np

import matplotlib.pyplot as plt

# Define the function
def f(x):
	return (x - 2) * (x - 4)

# Generate x values
x = np.linspace(-2, 6, 400)

# Generate y values
y = f(x)

# Create the plot
plt.plot(x, y, label='(x-2)(x-4)')

# Add labels and title
plt.xlabel('x')
plt.ylabel('y')
plt.title('Graph of the function (x-2)(x-4)')
plt.legend()

# Show the plot
plt.grid(True)
plt.show()
```
